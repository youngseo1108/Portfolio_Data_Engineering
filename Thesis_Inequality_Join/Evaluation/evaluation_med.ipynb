# Evaluation n = 5'000

```python
from src.generator import gen_intervals
from src.iejoin_func import iejoin
import pandas as pd
import numpy as np
import time, datetime
```

## 1. Case 0: no overlap

```python
# Case 0: no overlap
r0_med = pd.DataFrame(data=gen_intervals(5000, 2, 0.5, 0), columns=['P.B', 'P.E'])
r0_med['idx'] = r0_med.index
r0_med = r0_med[['idx', 'P.B', 'P.E']]
r0_med
```

### 1.1 r.idx > r'.idx AND r.B < r'.E

```python
start = time.time()
res = iejoin(r0_med, 'idx', 'P.B', r0_med, 'idx', 'P.E', 'gt', 'lt', 'idx', 'idx')
end = time.time()
print(str(datetime.timedelta(seconds=(end - start))))
print(len(res))
res
```

### 1.2 r.idx > r'.idx AND r.E > r'.B

```python
start = time.time()
res = iejoin(r0_med, 'idx', 'P.E', r0_med, 'idx', 'P.B', 'gt', 'gt', 'idx', 'idx')
end = time.time()
print(str(datetime.timedelta(seconds=(end - start))))
print(len(res))
res
```
