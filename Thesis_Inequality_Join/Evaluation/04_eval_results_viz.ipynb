## 2. Comparison among algorithms

```python
# Data -- comparison (n = 1000)
case0 = {
  "IEJoin": 3.19,
  "OMJ": 95.25,
  "BFJoin": 43.77
}

# "IEJoin-B<E'": 105.41,
case1 = {
  "IEJoin": 91.28,
  "OMJ": 55.62,
  "BFJoin": 38.48
}

case2_1 = {
  "IEJoin": 4.45,
  "OMJ": 222.90,
  "BFJoin": 130.76
}

# case2_2 = {
#   "IEJoin": 4.51,
#   "OMJ": 232.07,
#   "BFJoin": 136.16
# }

# "IEJoin-E>B'": 115.67,
case2_3 = {
  "IEJoin": 99.91,
  "OMJ": 79.50,
  "BFJoin": 50.69
}
```

```python
# Grouped data
labels = ['Case0', 'Case1', 'Case2 ($p=0.1$)', 'Case2 ($p>0.9$)']
iejoin_values = [case0['IEJoin'], case1['IEJoin'], case2_1['IEJoin'], case2_3['IEJoin']]
omj_values = [case0['OMJ'], case1['OMJ'], case2_1['OMJ'], case2_3['OMJ']]
bfjoin_values = [case0['BFJoin'], case1['BFJoin'], case2_1['BFJoin'], case2_3['BFJoin']]

x = np.arange(len(labels))  # the label locations
width = 0.25  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width, iejoin_values, width, label='IEJoin')
rects2 = ax.bar(x, omj_values, width, label='OMJ')
rects3 = ax.bar(x + width, bfjoin_values, width, label='BFJoin')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Runtime (seconds)', fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=12)
ax.tick_params(axis='y', labelsize=11)
ax.legend(fontsize=12)

fig.set_figwidth(8)

plt.show()
```
